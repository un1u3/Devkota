```python
from pathlib import Path

def merger(input_folder, output_file):
    input_folder = Path(input_folder)
    output_file = Path(output_file)
    output_file.parent.mkdir(parents=True, exist_ok=True)

    file_count = 0

    with open(output_file, "w", encoding="utf-8") as out:
        for filepath in input_folder.rglob("*.txt"):   # no sorting == less memory
            try:
                with open(filepath, "r", encoding="utf-8") as f:
                    empty = True
                    for line in f:                     # stream line-by-line
                        out.write(line)
                        empty = False

                    if not empty:
                        out.write("\n")               # separator
                        file_count += 1

                        if file_count % 1000 == 0:
                            print(f"Merged {file_count} files")

            except Exception as e:
                print(f"Skipped {filepath}: {e}")

    print(f"Done. Total files merged: {file_count}")
```

```python
merger(input_folder="../data/rawdata",
       output_file="../data/preprocessed/input.txt")
```
